```python
import pandas as pd
import numpy as np
import seaborn as sns
from tqdm import tqdm
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ligands_14_df = pd.read_pickle("14-ligands-dream35.pkl")
ligands_14_df_embeddings = ligands_14_df.drop(columns=["smiles"]).values
ligands_14_df_embeddings
```

```python
molformer_embeddings = pd.read_pickle("dream35-molformer-emb.pkl")
molformer_embeddings = molformer_embeddings.drop(columns=["smiles"]).values
molformer_embeddings
```

```python
distance_0 = cosine_similarity(ligands_14_df_embeddings, molformer_embeddings)
distance_0.shape
```

```python
distances = pd.DataFrame(distance_0)
distances
```

```python
top_5_idx = []
top_5_vals = []
for idx, row in distances.iterrows():
    top_5_idx.append(row.sort_values(ascending=False).head(5).index.values)
    top_5_vals.append(row.sort_values(ascending=False).head(5).values)
top_5_idx = np.array(top_5_idx)
top_5_vals = np.array(top_5_vals)
```

```python
top_5_idx, 14*5
```

```python
top_5_idx_flat = top_5_idx.flatten()
top_5_idx_flat = np.unique(top_5_idx_flat)
top_5_idx_flat.shape
```

```python
top_5_vals, top_5_idx_flat
```

```python
distances_max = distances.max(axis=0)
distances_max
```

```python
sel_50_distances = distances_max[top_5_idx_flat].sort_values(ascending=False).head(50)
sel_50_distances
```

```python
probs = pd.read_csv("final-pred-Subtask-2.csv")
probs["max_distance"] = distances_max
probs
```

```python
probs["Sel_50_distances"] = 0
# sel_50_distances.index is 1
probs.loc[sel_50_distances.index, "Sel_50_distances"] = 1
probs
```

```python
probs[["Sel_50","Sel_50_distances"]].value_counts()
```

```python
probs["Score_log10"] = -np.log10(probs["Score"])
probs
```

```python
sns.scatterplot(data=probs, x="Score_log10", y="max_distance")
```

```python
sns.scatterplot(data=probs[probs["Sel_50_distances"]==1], x="Score_log10", y="max_distance")
```

```python
# Save results
probs_ = probs[["RandomID", "Sel_50_distances", "max_distance"]]
probs_ = probs_.rename(columns={"Sel_50_distances": "Sel_50", "max_distance": "Score"})
probs_.to_csv("step-2-closeness.csv", index=False)
```
